```python

# Import OpenSeesPy
from openseespy.opensees import *
import matplotlib.pyplot as plt
import opsvis as opsv
import os
no_of_trials = 10
itr_no = 1
n_E = 1
# Clear any existing model
for i in range(1, no_of_trials):

  wipe()

  # Create a 2D model
  model('basic', '-ndm', 2, '-ndf', 3)

  # Beam geometry
  beamLength = 3.0  # total length in meters
  numElements = n_E
  elementLength = beamLength / numElements

  # Material and section properties
  E = 210e9          # Young's Modulus (Pa)
  A = 0.01          # Cross-sectional Area (m^2)
  I = 8.33333e-6        # Moment of Inertia (m^4)

  # Create nodes
  for i in range(numElements + 1):
      nodeTag = i + 1
      x = i * elementLength
      node(nodeTag, x, 0.0)
  print("NodeTag", nodeTag)
  # Fix the left end node
  fix(1, 1, 1, 1)  # Node 1 fixed in x, y, rotation

  # Define geometric transformation
  geomTransf('Linear', 1)

  # Define Elastic Section
  # We'll use elasticBeamColumn elements
  for i in range(1, numElements + 1):
      element('elasticBeamColumn', i, i, i+1, A, E, I, 1)

  # Apply point load at the rightmost node (node 4)
  #load(4, 0.0, -100.0, 0.0)  # 100N downward

  # Define analysis
  timeSeries('Linear', 1)
  pattern('Plain', 1, 1)

  # (Re-apply the load inside the load pattern)
  load(nodeTag, 0.0, -100.0, 0.0)

  # Constraints, DOF numbering, solution algorithm
  constraints('Plain')
  numberer('RCM')
  system('BandGeneral')
  test('NormDispIncr', 1.0e-6, 10)
  algorithm('Newton')
  integrator('LoadControl', 1.0)
  analysis('Static')

  # Run analysis
  analyze(1)
  reactions()
  # Print nodal displacements
  print("\nNodal Displacements:")
  for i in range(1, numElements + 2):
      dispX = nodeDisp(i, 1)
      dispY = nodeDisp(i, 2)
      rotation = nodeDisp(i, 3)
      print(f"Node {i}: UX={dispX:.6e} m, UY={dispY:.6e} m, RZ={rotation:.6e} rad")
  rx = nodeReaction(1, 1)
  ry = nodeReaction(1, 2)
  rz = nodeReaction(1, 3)
  print(f"Node {i}: RX={rx:.6e} N, RY={ry:.6e} N, MZ={rz:.6e} Nm")

  fig1, ax1 = plt.subplots(figsize=(40, 8))
  opsv.plot_model(node_labels=1, element_labels=1, ax = ax1)
  plt.title("Cantilever ElasticBeamColumn Element Model")

  fig3, ax3 = plt.subplots(figsize=(40, 8))
  opsv.plot_defo(ax = ax3)
  plt.title("Deformed Shape")
  plt.grid(True)
  plt.show()

  # Clean up
  itr_no+=1
  n_E+=1
  #folder = f'drive/MyDrive/Cantillever_1D_Diagrams/Iteration_{itr_no}'
  #os.makedirs(folder, exist_ok=True)
  #fig1.savefig(os.path.join(folder, f'Model_itr_{itr_no}.pdf'))
  #fig3.savefig(os.path.join(folder, f'Defo_itr_{itr_no}.pdf'))
  wipe()
```
